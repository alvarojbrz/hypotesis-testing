```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leitura dos dados

# Idade aposentadoria
dados_idade_aposentadoria = pd.read_csv("dados_idade_aposentadoria.csv")

# Tempo de vida de uma lâmpada
dados_vida_lampada = pd.read_csv("dados_vida_lampada.csv")

# Altura dos funcionários
dados_alturas = pd.read_csv("dados_alturas.csv")
```

```python
# Distribuições
plt.figure(figsize=(20,5))
plt.subplot(131)
plt.title('Distribuição de Idade de Aposentadoria')
plt.xlabel('Idade de Aposentadoria')
plt.ylabel('Frequência')
plt.hist(dados_idade_aposentadoria, bins=30, alpha=0.7, color='blue')

plt.subplot(132)
plt.hist(dados_vida_lampada, bins=30, alpha=0.7, color='gray')
plt.title('Tempo de Vida de uma Lâmpada')
plt.xlabel('Tempo de Vida (horas)')

plt.subplot(133)
plt.hist(dados_alturas, bins=30, alpha=0.7, color='brown')
plt.title('Alturas dos Funcionários')
plt.xlabel('Altura (cm)')

plt.tight_layout()
plt.show()
```

# Reamostragens

```python
# tamanho da amostra (n amostras)
n = 100

# quantidade de amostras 
qtd = 10000

# função de reamostragem e cálculo de médias
def reamostagem_medias(dados, coluna, n, qtd):
    medias = [dados[coluna].sample(n, replace=True).mean()for _ in range(qtd)]
    return medias
```

```python
# Realizando reamostragens
media_idade = reamostagem_medias(dados_idade_aposentadoria, 'idade', n, qtd)
media_duracao = reamostagem_medias(dados_vida_lampada, 'duracao', n, qtd)
media_altura = reamostagem_medias(dados_alturas, 'alturas', n, qtd)
```
